```python
vendedores = ['Ana', 'Luis', 'Marta', 'Sergio']
vendedores.append('Camila')


meses = []
ventas = []

cantidad_meses = int(input('Ingrese el número de meses a registrar: '))
for i in range(cantidad_meses):
    
    mes = input('Ingrese el nombre del mes: ')
    venta = float(input(f'Ingrese el total de ventas en {mes}: '))
    # Guardar los datos
    meses.append(mes)
    ventas.append(venta)

n = len(meses)
print('\nRegistro de Ventas Mensuales:')
for i in range(n):
    print(f'{meses[i]}: ${ventas[i]:,.2f}')
```

```python
#Ejercicio 3

nombre = input("Ingrese el nombre del estudiante: ")
cedula = input("Ingrese la cédula del estudiante: ")

examen1 = float(input("Ingrese la nota del primer examen: "))
examen2 = float(input("Ingrese la nota del segundo examen: "))
examen3 = float(input("Ingrese la nota del tercer examen: "))
seguimiento = float(input("Ingrese la nota de seguimiento del estudiante: "))

nota_final = examen1*0.20 + examen2*0.15 + examen3*0.40 + seguimiento*0.25

print(f"El nombre del estudiante es: {nombre}")
print(f"La cédula del estudiante es: {cedula}")
print(f"y su nota final en el curso es de: {nota_final}")
```
